```python
Caramelos= int(input("Ingresa el total de caramelos: "))
Cantidad_de_niños= int(input("Ingresa la cantidad de niños: "))
Caramelos_recibidos_por_cada_criatura= Caramelos/Cantidad_de_niños
print(f"Recibe cada niño {Caramelos_recibidos_por_cada_criatura}")
cuantos_sobran= Caramelos%Cantidad_de_niños
print(f"Sobran {cuantos_sobran} caramelos")
```
